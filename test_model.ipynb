```python
from db.mongo import get_all_invoices
from ml.predict import predict_match

transaction = {
    "sender_name": "Jan Kowalski",
    "amount": 1230.00,
    "currency": "PLN",
    "transfer_date": "2024-04-05",
    "description": "FV/2024/04/001 - op≈Çata"
}
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
async def test_matching():

    invoices = await get_all_invoices()
    results = []

    for invoice in invoices:
        try:
            prediction, prob = await predict_match(invoice, transaction)
            results.append((invoice["number"], prediction, prob))
        except Exception as e:
            print(f"B≈ÇƒÖd przy fakturze {invoice.get('number')}: {e}")

    for number, pred, prob in results:
        print(f"Faktura: {number}, Dopasowanie: {pred}, Prawdopodobie≈Ñstwo: {prob:.2f}")

    # (Opcjonalnie) Najbardziej prawdopodobna faktura:
    best = max(results, key=lambda x: x[2], default=None)
    if best:
        print(f"\nüîç Najlepsze dopasowanie: Faktura {best[0]} z prawdopodobie≈Ñstwem {best[2]:.2f}")
await test_matching()
```
